```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\Shalaby\Downloads\SMSSpamCollection.txt",sep='\t',names=['label','message'])
```

```python
df.head()
```

```python
## data cleaning
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
portstemmer = PorterStemmer()
lemmatizer  = WordNetLemmatizer()
```

```python
## remove special characters,lowercase all words,splitting into words then steming these words as long as there are not 
## in stopwords then join them again to have the words after preprocessing 
corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df["message"][i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if word not in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
## output feature
df.loc[df['label']=='ham',['label']] = 1
df.loc[df['label']=='spam',['label']] = 0
Y = df['label'].astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test = train_test_split(corpus,Y,test_size=0.2,random_state=42)
```

```python
##create bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500,ngram_range=(1,2),binary=True) ## take the top 2500 frequent words in the data
```

```python
X_train = cv.fit_transform(X_train).toarray()
X_test  = cv.transform(X_test).toarray()
```

```python
cv.vocabulary_
```

```python
## works very well with sparse data
from sklearn.naive_bayes import MultinomialNB
spam_ham_detector = MultinomialNB()
```

```python
spam_ham_detector.fit(X_train,Y_train)
```

```python
y_pred  = spam_ham_detector.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(Y_test,y_pred)
```

```python
print(classification_report(Y_test,y_pred))
```
